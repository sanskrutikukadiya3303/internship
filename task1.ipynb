```python
import pandas as pd
```

### **import dataset**

```python
df=pd.read_csv('/content/StudentsPerformance.csv')
df.head()
```

## **1. What is the average math, reading, and writing score overall?**



```python
df[['math score', 'reading score', 'writing score']].mean()
```

# 2. What is the average score of male vs female students?

```python
df.groupby('gender')[['math score', 'reading score', 'writing score']].mean()
```

# 3. Which parental education level is associated with the highest average scores?

```python
df[(df['math score'] > 90) & (df['reading score'] > 90) & (df['writing score'] > 90)].shape[0]
```

# 4. What is the distribution of lunch types and how does it relate to average scores?

```python
df.groupby('lunch')[['math score', 'reading score', 'writing score']].mean()
```

# 5. Create a bar plot of average scores by gender.

```python
import seaborn as sns
import matplotlib.pyplot as plt

avg_scores = df.groupby('gender')[['math score', 'reading score', 'writing score']].mean().reset_index()
avg_scores.plot(x='gender', kind='bar')
plt.title('Average Scores by Gender')
plt.ylabel('Score')
plt.show()
```

# 6.Create a boxplot showing math score distribution across different parental education levels.

```python
sns.boxplot(x='parental level of education', y='math score', data=df)
plt.xticks(rotation=45)
plt.title('Math Score by Parental Education Level')
plt.show()
```
